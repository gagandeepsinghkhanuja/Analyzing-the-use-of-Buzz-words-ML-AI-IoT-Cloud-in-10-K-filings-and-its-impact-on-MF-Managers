# Scraping S&P500 companies

```python
url='https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'### wikipedia url
```

```python
###using selenium to scrape data
from bs4 import BeautifulSoup as bs

from selenium import webdriver

path_to_chromedriver = 'C:\\Users\\swapn\\Desktop\\chromedriver_win32_2.27\\chromedriver.exe' # change path as needed
browser = webdriver.Chrome(executable_path = path_to_chromedriver)


browser.get(url) # Start the browser and open 'url'
assert 'List of S&P 500 companies - Wikipedia' in browser.title # Wait for the page to load
html = browser.page_source # Get the html of the page
browser.quit() # Close the browser
soup = bs(html, 'html.parser')


```

```python
##reading table from html
import pandas as pd
table=soup.find('table',class_="wikitable sortable jquery-tablesorter")
rows=[]
for row in table.find_all('tr'):
    rows.append([val.text.encode('utf8') for val in row.find_all('td')])
   

```

```python
rows
```

```python
##exporting csv
import csv

with open("SP_500.csv", "wb") as f:
    writer = csv.writer(f)
    writer.writerows(rows)
```
