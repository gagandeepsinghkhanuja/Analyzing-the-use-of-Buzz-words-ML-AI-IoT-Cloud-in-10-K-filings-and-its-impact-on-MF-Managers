```python
############Preparing Profit,StockPrice,ROE Variables

##Reading FIancial Data and transforming it
financial=pd.read_csv('Financials.csv')
financial['Ticker']=financial['Ticker'].map(lambda x: x.strip().lower())##reducing all characters to lower case
financial.head()
financial.drop(['0','Name'],axis=1,inplace=True)##dropping unnecessary columns
financial.columns=['Ticker', 'Profit_2012', 'Profit_2013',
       'Profit_2014', 'Profit_2015', 'Profit_2016',
       'Roe_2012', 'Roe_2013', 'Roe_2014', 'Roe_2015', 'Roe_2016',
       'Price_2012', 'Price_2013', 'Price_2014',
       'Price_2015', 'Price_2016']
financial.columns.values
financial.drop(['Profit_2012', 'Price_2012','Roe_2012'],inplace=True,axis=1)


##transforming data from wide to long
tidy_abs= pd.melt(financial, 
               id_vars='Ticker', 
               value_vars=[ 'Profit_2013', 'Profit_2014',
       'Profit_2015', 'Profit_2016', 'Roe_2013', 'Roe_2014',
       'Roe_2015', 'Roe_2016', 'Price_2013', 'Price_2014',
       'Price_2015', 'Price_2016'])

tidy_abs['Year']=tidy_abs['variable'].map(lambda x: x[x.find('_')+1:].strip())
tidy_abs['variable']=tidy_abs['variable'].map(lambda x: x[:x.find('_')].strip())
tidy_abs1= pd.DataFrame(tidy_abs.pivot_table(index=['Ticker','Year'], columns='variable', values='value'))##pivoting on Ticker,Year
flattened1 = pd.DataFrame(tidy_abs1.to_records())
flattened1.head()
```

```python
############Preparing Profit_change%,StockPrice_change%,ROE Variables_change%


financial=pd.read_csv('Financials.csv')
financial['Ticker']=financial['Ticker'].map(lambda x: x.strip().lower())
financial.head()
financial.drop(['0','Name'],axis=1,inplace=True)
financial.columns=['Ticker', 'Profit_2012', 'Profit_2013',
       'Profit_2014', 'Profit_2015', 'Profit_2016',
       'Roe_2012', 'Roe_2013', 'Roe_2014', 'Roe_2015', 'Roe_2016',
       'Price_2012', 'Price_2013', 'Price_2014',
       'Price_2015', 'Price_2016']




##tAKING RELATIVE CHANGE WRT TO LAST YEAR
financial['ProfPerc_2013']=financial['Profit_2013']/(financial['Profit_2012']+10**-6)-1##adding small number to denominator
financial['ProfPerc_2014']=financial['Profit_2014']/(financial['Profit_2013']+10**-6)-1
financial['ProfPerc_2015']=financial['Profit_2015']/(financial['Profit_2014']+10**-6)-1
financial['ProfPerc_2016']=financial['Profit_2016']/(financial['Profit_2015']+10**-6)-1

financial['RoePerc_2013']=financial['Roe_2013']/(financial['Roe_2012']+10**-6)-1
financial['RoePerc_2014']=financial['Roe_2014']/(financial['Roe_2013']+10**-6)-1
financial['RoePerc_2015']=financial['Roe_2015']/(financial['Roe_2014']+10**-6)-1
financial['RoePerc_2016']=financial['Roe_2016']/(financial['Roe_2015']+10**-6)-1

financial['PricePerc_2013']=financial['Price_2013']/(financial['Price_2012']+10**-6)-1
financial['PricePerc_2014']=financial['Price_2014']/(financial['Price_2013']+10**-6)-1
financial['PricePerc_2015']=financial['Price_2015']/(financial['Price_2014']+10**-6)-1
financial['PricePerc_2016']=financial['Price_2016']/(financial['Price_2015']+10**-6)-1


##dropping unnecessary variables
financial.drop(['Profit_2012', 'Profit_2013',
       'Profit_2014', 'Profit_2015', 'Profit_2016', 
                'Roe_2012', 'Roe_2013', 'Roe_2014', 'Roe_2015', 'Roe_2016',
       'Price_2012', 'Price_2013', 'Price_2014',
       'Price_2015', 'Price_2016'],inplace=True,axis=1)


###preparing wide to long data
tidy= pd.melt(financial, 
               id_vars='Ticker', 
               value_vars=['RoePerc_2013', 'RoePerc_2014', 'RoePerc_2015', 'RoePerc_2016',
       'ProfPerc_2013', 'ProfPerc_2014', 'ProfPerc_2015', 'ProfPerc_2016',
       'PricePerc_2013', 'PricePerc_2014', 'PricePerc_2015',
       'PricePerc_2016'])


###preparing new variables
tidy['Year']=tidy['variable'].map(lambda x: x[x.find('_')+1:x.find('_')+5])
tidy['variable']=tidy['variable'].map(lambda x: x[:x.find('_')])




###pivoting on ticker Year

tidy1= pd.DataFrame(tidy.pivot_table(index=['Ticker','Year'], columns='variable', values='value'))
flattened = pd.DataFrame(tidy1.to_records())
flattened.head()
```
